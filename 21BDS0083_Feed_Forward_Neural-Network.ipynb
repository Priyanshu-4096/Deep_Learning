```python
# I am creating a Sigmoid function
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
# Use the above function to estimate scores of different values of x
x = 1.0
print('21BDS0083 Sigmoid score of (%.1f): %.1f' % (x, sigmoid(x)))

x = 0.0
print('21BDS0083 Sigmoid score of (%.1f): %.1f' % (x, sigmoid(x)))

x = -1.0
print('21BDS0083 Sigmoid score of (%.1f): %.1f' % (x, sigmoid(x)))
```

```python
# Plot distribution of sigmoid scores over diffrent values of x [-ve to +ve]
import matplotlib.pyplot as plt    # you would need to implement matplotlib "pip install matplotlib"
x = np.linspace(-10, 10, 51)       # returns number spaces evenly w.r.t interval 25 -ve evenly spaced values and 25 +ve evenly spaced values.
print(x)

# find sigmoid values of the elements in the vector x
val = sigmoid(x)            # val is also a vector: val[i] is the sigmoid value of x[i]
plt.xlabel("21BDS0083 x") 
plt.ylabel("21BDS0083 Sigmoid(x)")  
plt.plot(x, val) 
plt.grid()
plt.show()
```

```python
def relu(x):
    return np.maximum(0.0,x)
```

```python
x = 1.0
print('21BDS0083 ReLU score of (%.1f): %.1f' % (x, relu(x)))

x = 0.0
print('21BDS0083 ReLU score of (%.1f): %.1f' % (x, relu(x)))

x = -1.0
print('21BDS0083 ReLU score of (%.1f): %.1f' % (x, relu(x)))
```

```python
def leaky_relu(x):
    return np.maximum(0.05*x,x)
```

```python
x = 1.0
print('21BDS0083 Leaky ReLU score of (%.2f): %.2f' % (x, leaky_relu(x)))

x = 0.0
print('21BDS0083 Leaky ReLU score of (%.2f): %.2f' % (x, leaky_relu(x)))

x = -1.0
print('21BDS0083 Leaky ReLU score of (%.2f): %.2f' % (x, leaky_relu(x)))
```

```python
import numpy as np

D = np.array([[1,2,3],
              [1,0,2],
              [0,1,4],
              [2,1,4]])

# Initialize Weight Matrices
np.random.seed(1)
W = np.random.random((3,4))
print("21BDS0083 W weight matrix:")
print(W)

print("21BDS0083 V weight matrix:")
V = np.random.random((4,2))
print(V)

for iteration in range(1):
    iLayer = D
    
    hP = np.dot(iLayer,W)        # Hidden Layer
    hLayer = 1/(1+np.exp(-hP))
    
    oP = np.dot(hLayer,V)        # Output Layer
    oLayer = 1/(1+np.exp(-oP))

print("21BDS0083 Input :")
print(D)  # Changed from 'training' to 'D'
print("21BDS0083 Predicted Output: ")
print(oLayer)
```
